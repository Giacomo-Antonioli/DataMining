This dataset describes different regions of different countries in different parts of the World and the used currency in these region. Each region is characterized by an unique code. 


```python
counter=0
for x in np.unique(geography_psd['continent']):
    print(x)
    tmp=geography_psd.loc[geography_psd['continent']==x]
    for y in np.unique(tmp['country']):
        print('\t'+y)
        tmp1=geography_psd.loc[ geography_psd['country']==y]
        for z in tmp1['region']:
                counter+=1
                print('\t\t'+z)
print("Unique regions: "+str(counter))
```

Each region is uniquely identified by a name and knowing that there are also 75 unique code identifiers we can drop the first coloumn of the table. We can consider this coloumn as a row counter, but having the same number of unique codes identifiers we can drop it. Also it some information that we can still retriver in other ways and we dont need to store it, because its dosent help us in any way with the analysis of the dataset.

```python
geography_psd=geography_psd.drop(columns=['Unnamed: 0'])
geography_psd.info()
```
